# Predicting Houses Prices using Linear Regression Algorithm

In this project, I aim to predict house prices using multiple features, including area, neighborhood, year built, heating, electrical systems, and several others. To enhance model performance, I applied standard scaling and employed Principal Component Analysis (PCA) for dimensionality reduction.

## Importing Libraries

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

## Data Preprocessing

#### - Drop the "Id" column and any features that are missing more than 40% of their values.

```python
df = df.drop('Id', axis=1)
```

```python
# check for missing data
nulls = df.isnull().sum()

# put the missing data in a dataframe
nulls_df = pd.DataFrame({'column': nulls.index, 'missing_data': nulls.values})
nulls_df = nulls_df[nulls_df['missing_data']!= 0]

```

```python
# drop the columns missing more than 40% of their values
treshold = (40 * len(df))/100

cols_to_drop = nulls_df.loc[nulls_df['missing_data']> treshold, 'column']
df = df.drop(cols_to_drop, axis=1)
```

```python
df.head()
```

#### -For numerical columns, fill in any missing data with the median value

```python
 # define the numerical features 
numerical_features= df.select_dtypes(include=np.number)

# fill the missing data with the median value 
for column in list(numerical_features.columns):
    df[column] = df[column].fillna(df[column].median())  
    
```

#### -For categorical columns, fill in any missing data with the most common value (mode)

```python
# define the categorical features 
categorical_features = df.select_dtypes(include='object')

# fill in the missing data with the mode value 
for column in list(categorical_features.columns):
    df[column] = df[column].fillna(df[column].mode())
```

#### -Apply dummy encoding to categorical columns 

```python
# Convert the categorical columns to dummy variables

df = pd.get_dummies(df, columns = list(categorical_features.columns), drop_first=True)


# convert the data type of dummy variables from booleen to integer
dummy_cols = df.select_dtypes(include='bool').columns

df[dummy_cols] = df[dummy_cols].astype(int)


# check the conversion of categorical columns to dummy variables
df.head()
```

#### -Split the data into a training and test set, where the SalePrice column is the target:

```python

# define the dependant (target variable) and the independant variables
X = df.drop('SalePrice', axis=1)
y = df['SalePrice']

# split the data into a training and test set
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=10)
```

#### - Scaling and applying PCA to the data 

```python

# initialize the StandardScaler
scaler = StandardScaler()

# standardize the training features
X_train_scaled = scaler.fit_transform(X_train)

# apply PCA while retaining 90% of the variance
pca = PCA(n_components = 0.90)
X_train_pca = pca.fit_transform(X_train_scaled)

# check the number of components retained
n_components_retained = pca.n_components
print(f"Number of components retained: {n_components_retained}")
```

```python
# print the Original number of features in the training data set
print("Original number of features:", X_train_scaled.shape[1])

# # print the number of features in PCA-transformed matrix
print("Reduced number of features:", X_train_pca.shape[1])
```

```python
# Scale and apply PCA to the test data:
```

```python
# standardize the testing data
X_test_scaled = scaler.transform(X_test)

# Transform but DO NOT fit the test features with the PCA
X_test_pca = pca.transform(X_test_scaled)
```

## Data Modeling and Evaluation

```python
# create the linear regression model
lr = LinearRegression()

# train the model on the training transformed data
lr.fit(X_train_pca, y_train)
```

```python
# make predictions on the test PCA transformed data
y_test_pca_pred = lr.predict(X_test_pca )

# calculate R2 and RMSE
r2_pca_test = r2_score(y_test_pca_pred, y_test)
rmse_pca_test = np.sqrt(mean_squared_error(y_test_pca_pred, y_test))

# print the metrics R2 and RMSE 
print("\nMetrics after PCA transformation on scaled test data:")
print(f"R2: {r2_pca_test:.2f}")
print(f"RMSE: {rmse_pca_test:.2f}")
```
