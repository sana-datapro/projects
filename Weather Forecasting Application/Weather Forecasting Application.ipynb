```python
import requests


def get_weather_zip(zip, unit):  # create a function to do a geo lookup and then a weather lookup by zip code.
    api_key = 'a0ed3df04ddb1a971eed18eba2ecf913'
    lang = "en"
    try:
        #  call api including the zip code input by user to get latitude &longitude.
       geo_link = f"http://api.openweathermap.org/geo/1.0/zip?zip={zip},US&appid={api_key}&lang={lang}"
       res = requests.get(geo_link)
       data = res.json()
       lat = data['lat']
       lon = data['lon']
    #  call api to get weather data using latitude & longitude.
       weather_link = f"https://api.openweathermap.org/data/2.5/weather?lat={lat}&lon={lon}&appid={api_key}&units={unit}&lang={lang}"
       #complete_weather_link = weather_link + "US&appid=" + api_key + "&" + unit + "&lang=" + lang
       response = requests.get(weather_link)  # Use the Requests library to request data from the webservice
       weather_data = response.json()
       display_data(weather_data, unit)
    except ConnectionError:
        print("There was a connection issue.")
```

```python
def get_weather_city(city, unit):  # create a function to do a geo lookup and then a weather lookup by city name.
    api_key = 'a0ed3df04ddb1a971eed18eba2ecf913'
    try:
        # call api including the city name input by user to get latitude &longitude & state.
        lang = "en"
        geo_link = f"http://api.openweathermap.org/geo/1.0/direct?q={city},US&limit=1&appid={api_key}"
        #geo_link = 'http://api.openweathermap.org/geo/1.0/direct?q={},US, &limit=1'.format(city)
        #complete_geo_link = geo_link + "&" + "appid=" + api_key
        res = requests.get(geo_link)
        data = res.json()
        lat = data[0]['lat']
        lon = data[0]['lon']
        state = data[0]['state']
        # call api to get weather data using latitude & longitude & state.
        weather_link = f"https://api.openweathermap.org/data/2.5/weather?lat={lat}&lon={lon}&appid={api_key}&units={unit}&lang={lang}"
        #weather_link = 'https://api.openweathermap.org/data/2.5/weather?lat={}&lon={}'.format(lat, lon)
        #complete_weather_link = weather_link + "&appid=" + api_key + "&" + unit + "&" + state
        response = requests.get(weather_link)
        weather_data = response.json()
        display_data(weather_data, unit)
    except ConnectionError:
        print("There is a connection issue.")
```

```python
def validate_zip(zip):  # create a function to validate if the zip code input is a number
    try:
        int(zip)
        return True
    except ValueError:
        print("You didn't enter a valid number, you entered", zip)
        return False
```

```python
def validate_city(city):  # create a function to validate if the city name input is alphabetic
    if not city.isdigit():
        return True
    else:
        print("Invalid city input, you entered", city)
        return False
```

```python
def display_data(weather_data, unit):  # create a function to show the current weather data to the user.
    location = weather_data['name']
    current_temp = weather_data['main']['temp']
    temp_feels_like = weather_data['main']['feels_like']
    low_temperature = weather_data['main']['temp_min']
    high_temp = weather_data['main']['temp_max']
    pressure = weather_data['main']['pressure']
    humidity = weather_data['main']['humidity']
    description = weather_data['weather'][0]['description']

    #  print the weather data requested, use if statements to print the temperature values in the unit chosen by user.

    print("------------------------------------------------------------------")

    print("Current weather conditions for {}: ".format(location))
    print("------------------------------------------------------------------")
    if unit.lower() == "metric":
        print("Current temperature :{} \N{DEGREE SIGN}C".format(current_temp))
        print("Temperature feels like {} \N{DEGREE SIGN}C".format(temp_feels_like))
        print("Low temperature:{} \N{DEGREE SIGN}C".format(low_temperature))
        print("High temperature:{} \N{DEGREE SIGN}C".format(high_temp))
    elif unit.lower() == "imperial":
        print("Current temperature :{}  \N{DEGREE SIGN} F".format(current_temp))
        print("Temperature feels like {} \N{DEGREE SIGN} F".format(temp_feels_like))
        print("Low temperature:{} \N{DEGREE SIGN} F".format(low_temperature))
        print("High temperature:{} \N{DEGREE SIGN} F".format(high_temp))
    else:
        print("Current temperature :{} K".format(current_temp))
        print("Temperature feels like {} K".format(temp_feels_like))
        print("Low temperature:{} K".format(low_temperature))
        print("High temperature:{} K".format(high_temp))

    print("Pressure: {} hPa".format(pressure))
    print("Humidity: {} %".format(humidity))
    print("Description: {}".format(description))


```

```python
def main():
    print("Welcome to our Weather lookup Program!")  # Welcoming the user.
    while True:
        # ask the user to input YES to proceed his request or NO to exit.
        user_input = input("Please enter YES if you want to know the current weather conditions or NO to exit:  ")
        #  if user input is NO, the user will quit the program.
        if user_input.upper() == "NO":
            print("Thank you for using our program.")
            break
        else:

            while user_input.upper() == "YES":
                # ask the user to input the search method he wants: by zip code or by city name.
                choice = input("Please choose one option: \n 1.Enter ZIP if you want to search for weather by zip code." 
                               "\n 2.Enter CITY if you want to search for weather by city name. \n ")
                if choice.upper() == "ZIP" or choice.upper() == "CITY":

                    # ask the user to input the temperature unit of measurement he wants to get his data.
                    unit_choice = input("Choose the temperature unit of measurement you want from the following: "       
                                        "\n 1. C for Celsius. \n 2. F for Fahrenheit. \n 3. K for kelvin.\n ")

                    if unit_choice.upper() == "F":
                        unit = "imperial"
                    elif unit_choice.upper() == "C":
                        unit = "metric"
                    elif unit_choice.upper() == "K":
                        unit = "standard"
                    else:
                        print(" Invalid unit, please enter F or C or K")
                    while unit_choice.upper() == "F" or unit_choice.upper() == "C" or unit_choice.upper() == "K":
                        # if unit choice is valid, process the user request
                        if choice.upper() == "ZIP":
                            zip = input("Enter the zip code: ")
                            if validate_zip(zip):
                                try:
                                    get_weather_zip(zip, unit)
                                    break
                                except Exception:
                                    print("There is no data related to this zip code, please enter a valid zip code.")

                        if choice.upper() == "CITY":
                            city = str(input("Enter your city name: "))
                            if validate_city(city):
                                try:
                                    get_weather_city(city, unit)
                                    break
                                except Exception:
                                    print("There are no data related to that city, enter a valid city name.")
                    break
                else:
                    print("Invalid choice input, please enter ZIP or CITY")


if __name__ == "__main__":
    main()

```
